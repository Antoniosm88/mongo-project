```python
from pymongo import MongoClient
from pymongo import ASCENDING, DESCENDING
import pandas as pd
import numpy as np
import re
import requests
import json
import os
from dotenv import load_dotenv
import googlemaps
import pprint
from googlemaps import convert
load_dotenv()
dbName = "companies"
mongodbURL = f"mongodb://localhost/{dbName}"
client = MongoClient(mongodbURL, connectTimeoutMS=2000, serverSelectionTimeoutMS=2000)
db = client.get_database()
cursor=db.list_collections()
```

Realizo una query para seleccionar del db las empresas que pertenecen al mismo sector.

```python
query = {
    "$or":[
        {"category_code":{"$eq":"web" }},
        {"category_code":{"$eq":"games_video" }},
        {"category_code":{"$eq":"ecommerce" }},
        {"category_code":{"$eq":"social" }},
        {"funding_rounds.investments":{"$gte":1000000 }}
        ]
}

cur = db.companies.find(query)
df =pd.DataFrame(list(cur))
df.head()
```

```python
offices = df.explode("offices").apply(lambda e: e.offices,axis=1,result_type="expand")
offices.head()

```

Miro en el database los estado con más oficinas, elijo NY.

```python
offices["state_code"].value_counts()
```

```python
#Elegimos New York
offices2 = offices.loc[offices["state_code"] == "NY"]
offices2 = offices2.dropna()
offices2 = offices2.reset_index()
offices2 = offices2.drop(columns = ["index"])
offices2
```

```python
nyork_offices = pd.concat([df[["name","_id"]], offices2], axis=1)
nyork_offices["state_code"].value_counts()
```

```python
#Quitar los "nan" de la columna "latitude" y "longitude" y que sea None
def transformToGeoPoint(s):
    if np.isnan(s.latitude) or np.isnan(s.longitude):
        return None
    return {
        "type":"Point",
        "coordinates":[s.longitude, s.latitude]
    }
#Añadimos la columna "geopoint"  con las la transformacion hecha.  
nyork_offices["geopoint"] = nyork_offices.apply(transformToGeoPoint, axis=1)
```

```python
def getData(query, page_token=None):
    url="https://maps.googleapis.com/maps/api/place/textsearch/json?"
    params= {"key":os.getenv("token2"), "query":query}
    if page_token:
        params["pageToken"]=page_token
    res= requests.get(url, params=params)
    res= res.json()
    results=res["results"]
    if "next_page_token" in res.keys():
        page_token = res["next_page_token"]
        results=[results, page_token]
    else:
        results=results,False
    return results
```

```python
data = pd.DataFrame(data_starbucks)
data.head()
```

```python
starbucks = []
for coffe in data["geometry"]:
    new_dict={}
    new_dict["Latitude"]= coffe['location']['lat']
    new_dict["Longitude"]= coffe['location']['lng']
    starbucks.append(new_dict)

starbucks = pd.DataFrame(starbucks)
starbucks.head()
```
